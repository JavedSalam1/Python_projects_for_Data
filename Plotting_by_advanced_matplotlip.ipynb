## Annotations

```python
batter=pd.read_csv('C:/Users/USER/Downloads/batter.csv')
```

```python
sample_df=batter.head(100).sample(25,random_state=5)
sample_df.head()
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])

plt.xlabel('Avg')
plt.ylabel('Strik_rate')

```

## Horizonta line and verticial line

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])

plt.xlabel('Avg')
plt.ylabel('Strik_rate')

```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[1].scatter(batter['avg'],batter['runs'])
ax[0].set_title('Avg vs Strike rate')

ax[0].set_ylabel('Strike rate')
ax[1].set_title('Avg vs Runs')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,6))
ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['avg'],color='red')
ax[1,1].hist(batter['runs'])
ax[0,0].set_title('Avg vs Strike rate')
ax[0,1].set_title('Avg vs Runs')
```

```python
#3D Scatte rplot
fig=plt.Figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(batter['runs'],batter['avg'],batter['strike_rate'])
ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('Strike rate')
ax.set_title("IPL Batsman Analysis")
```

```python
#3D surface plot
x=np.linspace(-10,10.,100)
y=np.linspace(-10,10.,100)
```

```python
xx,yy=np.meshgrid(x,y)
z=xx**2 + yy**2
z.shape
```

```python
fig=plt.Figure(figsize=(12,6))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.Figure(figsize=(12,6))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.Figure(figsize=(12,6))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.Figure(figsize=(12,6))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
